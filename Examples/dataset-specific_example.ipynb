```python
# import unioncom and numpy
import numpy as np
from unioncom import UnionCom

# integrate simulation 2 by UnionCom 
data1 = np.loadtxt("./simu2/domain1.txt")
data2 = np.loadtxt("./simu2/domain2.txt")

integrated_data = UnionCom.fit_transform([data1, data2], epoch_pd=50000, log_pd=10000)

# test label transfer accuracy
type1 = np.loadtxt("./simu2/type1.txt")
type2 = np.loadtxt("./simu2/type2.txt")
UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])

# visualization by PCA
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```

```python
import numpy as np
from unioncom import UnionCom

# integrate simulation 1 with specific data in MMD-MA paper by UnionCom 
data1 = np.loadtxt("./MMD/s1_mapped1.txt")
data2 = np.loadtxt("./MMD/s1_mapped2.txt")

type1 = np.loadtxt("./MMD/s1_type1.txt")
type2 = np.loadtxt("./MMD/s1_type2.txt")    
index1 = np.argwhere(type1==0).reshape(1,-1).flatten()
index2 = np.argwhere(type1==2).reshape(1,-1).flatten()
index = np.hstack((index1, index2))
type1 = type1[index]
data1 = data1[index]

integrated_data = UnionCom.fit_transform([data1, data2], log_pd=10000)

# test label transfer accuracy

UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])

# visualization by PCA
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```

```python
import numpy as np
from unioncom import UnionCom

# integrate scNMT data with specific data by UnionCom 
data1 = np.loadtxt("./scNMT/Eaccess.txt")
data2 = np.loadtxt("./scNMT/Emethy.txt")

type1 = np.loadtxt("./scNMT/type1.txt")
type2 = np.loadtxt("./scNMT/type2.txt")    
index1 = np.argwhere(type1==1).reshape(1,-1).flatten()
index2 = np.argwhere(type1==2).reshape(1,-1).flatten()
index = np.hstack((index1, index2))
type1 = type1[index]
data1 = data1[index]

integrated_data = UnionCom.fit_transform([data1, data2], epsilon=0.0001, beta=0.1, log_pd=10000)

# test label transfer accuracy

UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])

# visualization by PCA
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```
