```python
# import unioncom and numpy
from unioncom import UnionCom
import numpy as np
```

```python
# load simu1 data
data1 = np.loadtxt("./simu1/domain1.txt")
data2 = np.loadtxt("./simu1/domain2.txt")
```

```python
# integrate data by UnionCom
integrated_data = UnionCom.fit_transform([data1, data2])
```

```python
# test label transfer accurcay
type1 = np.loadtxt("./simu1/type1.txt")
type2 = np.loadtxt("./simu1/type2.txt")
UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])
```

```python
# PCA visulaization
type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```

```python
# import unioncom and numpy
from unioncom import UnionCom
import numpy as np
# load simu2 data
data1 = np.loadtxt("./simu2/domain1.txt")
data2 = np.loadtxt("./simu2/domain2.txt")
type1 = np.loadtxt("./simu2/type1.txt")
type2 = np.loadtxt("./simu2/type2.txt")
integrated_data = UnionCom.fit_transform([data1, data2], epoch_pd=50000, log_pd=10000)

UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])

type1 = type1.astype(np.int)
type2 = type2.astype(np.int)

UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```

```python
# use barycentric instead of tsne
from unioncom import UnionCom
import numpy as np
data1 = np.loadtxt("./simu1/domain1.txt")
data2 = np.loadtxt("./simu1/domain2.txt")

integrated_data = UnionCom.fit_transform([data1, data2], project='barycentric', log_pd=10000)

type1 = np.loadtxt("./simu1/type1.txt")
type2 = np.loadtxt("./simu1/type2.txt")
UnionCom.test_label_transfer_accuracy(integrated_data, [type1,type2])

type1 = type1.astype(np.int)
type2 = type2.astype(np.int)
UnionCom.PCA_visualize([data1,data2], integrated_data, [type1,type2])
```
